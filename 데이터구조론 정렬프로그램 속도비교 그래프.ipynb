```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
selection = [0.14171, 0.17589, 0.20548, 0.12952, 0.14931]
bubble = [0.81932, 0.00059, 1.14274, 0.63803, 0.75131]
insertion = [0.20307, 0.00119, 0.38898, 0.04957, 0.14731]
```

```python
from pylab import legend
from pylab import plot,show, axis,title,xlabel,ylabel
```

```python
data = [i for i in range(1,6)]
plot(data, selection, data, bubble, data, insertion, marker='o')
axis(xmin =0, ymin=0)
xlabel("data")
ylabel('Time')
legend(["Selection","Bubble","Insertion"], loc='best')
show()
```
